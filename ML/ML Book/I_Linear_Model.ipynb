```python
import numpy as np
import pandas as pd

# Example data for oecd_bli and gdp_per_capita
oecd_bli = pd.DataFrame({
    "Country": ["CountryA", "CountryB", "CountryC"],
    "Life satisfaction": [7.2, 6.5, 8.0]
})

gdp_per_capita = pd.DataFrame({
    "Country": ["CountryA", "CountryB", "CountryC"],
    "GDP per capita": [40000, 35000, 50000]
})

# Define the prepare_country_stats function
def prepare_country_stats(oecd_bli, gdp_per_capita):
    # Merge the two DataFrames on the 'Country' column
    country_stats = pd.merge(oecd_bli, gdp_per_capita, on="Country")
    
    # Return the relevant columns (Country, GDP per capita, and Life satisfaction)
    return country_stats[["Country", "GDP per capita", "Life satisfaction"]]

# Prepare the country stats
country_stats = prepare_country_stats(oecd_bli, gdp_per_capita)

# Extract X (GDP per capita) and y (Life satisfaction)
X = np.c_[country_stats["GDP per capita"]]
y = np.c_[country_stats["Life satisfaction"]]

# Print X and y
print("X (GDP per capita):\n", X)
print("y (Life satisfaction):\n", y)
```

```python
# Select a linear model
model = sklearn.linear_model.LinearRegression()
```

```python
# Train the model
model.fit(X, y)
```

```python
# Make a prediction for Cyprus
X_new = [[22587]]  # Cyprus' GDP per capita
print(model.predict(X_new)) # outputs [[ 5.96242338]]
```
