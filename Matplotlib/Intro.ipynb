# Font-Style

```python
import matplotlib.pyplot as plt

x = [15, 12, 23]
y = [10, 20, 30]

plt.plot(x, y)

plt.title("****** Ns-loni 1st Matplotlib Tutorial in 2025 ******")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')  # Corrected this line

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [15, 12, 23]
y = [10, 20, 30]

plt.plot(x,y)

plt.title("Ns-loni 1st Matplotlib Tutorial in 2*25", fontdict={'fontname':'comic Sans MS'})

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([10,11,12,23,24])
plt.yticks([10,22,24,36,38])

plt.show()
```

# Line Color

```python
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, label='2x', color = 'red')

plt.title("Ns-loni 1st Matplotlib Tutorial in 2*25", fontdict={'fontname':'comic Sans MS'})

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.show()
```

# Line width & Markersize markeredgecolor and more



```python
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, label='2x', color='green', linewidth = 1, marker='.',  markersize = 10, markeredgecolor = 'red')

plt.title("Ns-loni 1st Matplotlib Tutorial in 2*25", fontdict={'fontname':'comic Sans MS'})

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, label='2x', color='green', linewidth = 1, marker='.',  markersize = 10, markeredgecolor = 'red')


# 2nd Line

x2 = np.arange(0, 4, 0.5)

plt.plot(x2, x2**2, 'r')  # r = red color


plt.title("Ns-loni 1st Matplotlib Tutorial in 2*25", fontdict={'fontname':'comic Sans MS'})

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.show()
```

# Resize the your Notation


```python
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, label='2x', color='green', linewidth = 1, marker='.',  markersize = 10, markeredgecolor = 'red')


# 2nd Line

x2 = np.arange(0, 4, 0.5)

plt.plot(x2, x2**2, 'r')  # r = red color

# Resize the your Notation
plt.plot(x,y, 'b^--', label='2x')


plt.title("Ns-loni 1st Matplotlib Tutorial in 2*25", fontdict={'fontname':'comic Sans MS'})

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.show()
```
